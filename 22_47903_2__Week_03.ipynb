## Smart Home Automation System

In the **Smart Home System** when some approaches the door a camera sence and perform facial recognition. If recognize then door open, if not recognize then the owner notified to grant or deny access or remotely. If the door is tempered with forcefully an alarm is activate and notify as an emergency. The door can also open using a pin or physical key for added security and convenience

# Agent Description:

1. Objective: The aim of the agent is to ensure secure access to the home by unlocking the door for recognized faces or valid PINs, and locking the door once entry is detected to maintain safety.


2. List the sensors:
 > Motion Detector <br>
 > Camera <br>
 > Keypad <br>
 > Pressure Sensor <br>
 > Pad Lock


3. Sensors' values:
Motion -> {Motion Detected, Motion Not Detect}, <br>
CameraStatus -> {Camera ON, Camera OFF}, <br>
FaceStatus = -> {Face Recognized, Face Not Recognized}, <br>
PinEnterStatus = -> {Pin Enter Corrctly, Pin Enter Incorrectly}, <br>
PhysicalKeyStatus -> {Correct Physical key used, Incorrect Physical key used}, <br>
PeopleStatus -> {People Entered, People Exited}, <br>
OwnerDecision -> {Owner Granted, Owner Denied}


4. List the actions:
 > Camera ON,<br>
 Camera OFF, <br>
 Door ON, <br>
 Door OFF, <br>
 Alarm OFF, <br>
 Door Open, <br>
 Door Lock


5. Agent function: **perception --> action**
> (Motion Not Detected) --> Camera Off <br>
(Motion Detected) --> Camera On <br>
(Camera On, Face Recognized) --> Door Open <br>
(Camera On, Face Not Recognized) --> Keep Door Closed, Notify Owner<br>
(Camera On, Face Not Recognized, Owner Granted) --> Door Open <br>
(Camera On, Face Not Recognized, Owner Denied) --> Keep Door Closed <br>
(Motion Detected, Camera On, Face Recognized) --> Door Open <br>
(Motion Detected, Camera On, Face Recognized, People Entered) --> Door Closed <br>
(Motion Detected, Camera On, Face Not Recognized) --> Keep Door Closed, Notify Owner <br>
(Motion Detected, Camera On, Face Not Recognized, Owner Granted) --> Door Open <br>
(Motion Detected, Camera On, Face Not Recognized, Owner Denied) --> Keep Door Closed <br>
(Door Tampered) --> Notify Owner, Alarm System Activate <br>
(Door Tampered, Owner Alarm Deactivated) --> Alarm Off <br>
(PIN Entered Correctly) --> Door Open <br>
(PIN Entered Incorrectly) --> Keep Door Closed, Notify Owner <br>
(PIN Entered Incorrectly, Owner Granted) --> Door Open <br>
(PIN Entered Incorrectly, Owner Denied) --> Keep Door Closed <br>
(Correct Physical Key Used) --> Door Open <br>
(Incorrect Physical Key Used) --> Keep Door Closed, Notify Owner <br>
(Incorrect Physical Key Used, Owner Granted) --> Door Open <br>
(Incorrect Physical Key Used, Owner Denied) --> Keep Door Closed <br>
(People Entered) --> Lock Door <br>
(People Exited) --> Lock Door



# Table driven agent

```python
table = {
    ('Motion Not Detected',): 'Camera Off',
    ('Motion Detected',): 'Camera On',
    ('Camera On', 'Face Recognized'): 'Door Open',
    ('Camera On', 'Face Not Recognized'): ('Keep Door Closed', 'Notify Owner'),
    ('Camera On', 'Face Not Recognized', 'Owner Granted'): 'Door Open',
    ('Camera On', 'Face Not Recognized', 'Owner Denied'): 'Keep Door Closed',
    ('Motion Detected', 'Camera On', 'Face Recognized'): 'Door Open',
    ('Motion Detected', 'Camera On', 'Face Recognized', 'People Entered'): 'Door Closed',
    ('Motion Detected', 'Camera On', 'Face Not Recognized'): ('Keep Door Closed', 'Notify Owner'),
    ('Motion Detected', 'Camera On', 'Face Not Recognized', 'Owner Granted'): 'Door Open',
    ('Motion Detected', 'Camera On', 'Face Not Recognized', 'Owner Denied'): 'Keep Door Closed',
    ('Door Tampered',): ('Notify Owner', 'Alarm System Activate'),
    ('Door Tampered', 'Owner Alarm Deactivated'): 'Alarm Off',
    ('PIN Entered Correctly',): 'Door Open',
    ('PIN Entered Incorrectly',): ('Keep Door Closed', 'Notify Owner'),
    ('PIN Entered Incorrectly', 'Owner Granted'): 'Door Open',
    ('PIN Entered Incorrectly', 'Owner Denied'): 'Keep Door Closed',
    ('Correct Physical Key Used',): 'Door Open',
    ('Incorrect Physical Key Used',): ('Keep Door Closed', 'Notify Owner'),
    ('Incorrect Physical Key Used', 'Owner Granted'): 'Door Open',
    ('Incorrect Physical Key Used', 'Owner Denied'): 'Keep Door Closed',
    ('People Entered',): 'Lock Door',
    ('People Exited',): 'Lock Door'
}
```

```python
percepts = []  # to store percept history

def table_driven_agent(percept):
    print('Perception Received: ' + str(percept))
    percepts.append(percept)  # updating percept history
    action = table[percept]
    return action
```

```python
# lets simulate the agent
import random

# Simulating the first perception randomly
motion_detected = random.choice([True, False])
if motion_detected:
    face_recognized = random.choice([True, False])
    if face_recognized:
        percept = ('Motion Detected', 'Camera On', 'Face Recognized')
    else:
        owner_granted = random.choice([True, False])
        percept = ('Motion Detected', 'Camera On', 'Face Not Recognized')
        if owner_granted:
            percept += ('Owner Granted',)
        else:
            percept += ('Owner Denied',)
else:
    percept = ('Motion Not Detected',)

# To perceive environment repeatedly
while True:
    # The agent program is being called with current perception
    action = table_driven_agent(percept)
    print('Action Performed: ' + str(action))

    # User control to allow next perception
    cmd = input('Type x to exit! or any key to continue.')
    if cmd == 'x': break

    # Simulating the next perception based on action
    if 'Camera Off' in action:
        motion_detected = random.choice([True, False])
        percept = ('Motion Not Detected',) if not motion_detected else ('Motion Detected',)
    elif 'Camera On' in action:
        face_recognized = random.choice([True, False])
        if face_recognized:
            percept = ('Motion Detected', 'Camera On', 'Face Recognized')
        else:
            owner_granted = random.choice([True, False])
            percept = ('Motion Detected', 'Camera On', 'Face Not Recognized')
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
    elif 'Door Open' in action:
        people_entered = random.choice([True, False])
        if people_entered:
            percept = ('People Entered',)
        else:
            people_exited = random.choice([True, False])
            if people_exited:
                percept = ('People Exited',)
    elif 'Alarm System Activate' in action:
        owner_alarm_deactivated = random.choice([True, False])
        percept = ('Door Tampered',)
        if owner_alarm_deactivated:
            percept += ('Owner Alarm Deactivated',)
    elif 'Keep Door Locked' in action or 'Notify Owner' in action:
        pass

    # Simulating PIN or Physical Key entry
    pin_entered = random.choice([True, False])
    if pin_entered:
        pin_correct = random.choice([True, False])


        if pin_correct:
            percept = ('PIN Entered Correctly',)
        else:
            owner_granted = random.choice([True, False])
            percept = ('PIN Entered Incorrectly',)
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
    physical_key_used = random.choice([True, False])
    if physical_key_used:
        key_correct = random.choice([True, False])
        if key_correct:
            percept = ('Correct Physical Key Used',)
        else:
            owner_granted = random.choice([True, False])
            percept = ('Incorrect Physical Key Used',)
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
```

# Simple reflex agent

```python
def simple_reflex_agent(percept):
    print('Perception Received: ' + str(percept))

    # Extract relevant parts of the percept
    motion_detected = 'Motion Detected' in percept
    face_recognized = 'Face Recognized' in percept
    face_not_recognized = 'Face Not Recognized' in percept
    door_tampered = 'Door Tampered' in percept
    owner_granted = 'Owner Granted' in percept
    owner_denied = 'Owner Denied' in percept
    pin_entered_correctly = 'PIN Entered Correctly' in percept
    pin_entered_incorrectly = 'PIN Entered Incorrectly' in percept
    correct_key_used = 'Correct Physical Key Used' in percept
    incorrect_key_used = 'Incorrect Physical Key Used' in percept
    people_entered = 'People Entered' in percept
    people_exited = 'People Exited' in percept

    # Determine the action based on the percept
    if not motion_detected:
        action = 'Camera Off'
    elif motion_detected:
        action = 'Camera On'
        if face_recognized:
            action = 'Door Open'
        elif face_not_recognized:
            if owner_granted:
                action = 'Door Open'
            elif owner_denied:
                action = 'Keep Door Locked'
            else:
                action = ('Keep Door Locked', 'Notify Owner')

    elif door_tampered:
        if 'Owner Alarm Deactivated' in percept:
            action = 'Alarm Off'
        else:
            action = ('Notify Owner', 'Alarm System Activate')
    elif pin_entered_correctly:
        action = 'Door Open'
    elif pin_entered_incorrectly:
        if owner_granted:
            action = 'Door Open'
        elif owner_denied:
            action = 'Keep Door Locked'
        else:
            action = ('Keep Door Locked', 'Notify Owner')
    elif correct_key_used:
        action = 'Door Open'
    elif incorrect_key_used:
        if owner_granted:
            action = 'Door Open'
        elif owner_denied:
            action = 'Keep Door Locked'
        else:
            action = ('Keep Door Locked', 'Notify Owner')
    elif people_entered or people_exited:
        action = 'Lock Door'

    return action
```

```python
import random

# Simulating the first perception randomly
motion_detected = random.choice([True, False])
if motion_detected:
    face_recognized = random.choice([True, False])
    if face_recognized:
        percept = ('Motion Detected', 'Face Recognized')
    else:
        owner_granted = random.choice([True, False])
        percept = ('Motion Detected', 'Face Not Recognized')
        if owner_granted:
            percept += ('Owner Granted',)
        else:
            percept += ('Owner Denied',)
else:
    percept = ('Motion Not Detected',)

# To perceive environment repeatedly
while True:
    # The agent program is being called with current perception
    action = simple_reflex_agent(percept)
    print('Action Performed: ' + str(action))

    # User control to allow next perception
    cmd = input('Type x to exit! or any key to continue.')
    if cmd == 'x': break

    # Simulating the next perception based on action
    if 'Camera Off' in action:
        motion_detected = random.choice([True, False])
        percept = ('Motion Not Detected',) if not motion_detected else ('Motion Detected',)
    elif 'Camera On' in action:
        face_recognized = random.choice([True, False])
        if face_recognized:
            percept = ('Motion Detected', 'Face Recognized')
        else:
            owner_granted = random.choice([True, False])
            percept = ('Motion Detected', 'Face Not Recognized')
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
    elif 'Door Open' in action:
        people_entered = random.choice([True, False])
        if people_entered:
            percept = ('People Entered',)
        else:
            people_exited = random.choice([True, False])
            if people_exited:
                percept = ('People Exited',)
    elif 'Alarm System Activate' in action:
        owner_alarm_deactivated = random.choice([True, False])
        percept = ('Door Tampered',)
        if owner_alarm_deactivated:
            percept += ('Owner Alarm Deactivated',)
    elif 'Keep Door Closed' in action or 'Notify Owner' in action:
        # Keep the same state for the next round
        pass

    # Simulating PIN or Physical Key entry
    pin_entered = random.choice([True, False])
    if pin_entered:
        pin_correct = random.choice([True, False])
        if pin_correct:
            percept = ('PIN Entered Correctly',)
        else:
            owner_granted = random.choice([True, False])
            percept = ('PIN Entered Incorrectly',)
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
    physical_key_used = random.choice([True, False])
    if physical_key_used:
        key_correct = random.choice([True, False])
        if key_correct:
            percept = ('Correct Physical Key Used',)
        else:
            owner_granted = random.choice([True, False])
            percept = ('Incorrect Physical Key Used',)
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
```

# Model based agent

```python
# Variables to store available information about the world state
world_state = {
    'Motion': 'Unknown',
    'Camera': 'Off',
    'Face': 'Unknown',
    'Door': 'Closed',
    'Alarm': 'Off'
}

def update_state(action, percept, model):
    if 'Motion Detected' in percept:
        model['Motion'] = 'Detected'
        model['Camera'] = 'On'
    elif 'Motion Not Detected' in percept:
        model['Motion'] = 'Not Detected'
        model['Camera'] = 'Off'

    if 'Face Recognized' in percept:
        model['Face'] = 'Recognized'
    elif 'Face Not Recognized' in percept:
        model['Face'] = 'Not Recognized'
        if 'Owner Granted' in percept:
            model['Face'] = 'Granted'
        elif 'Owner Denied' in percept:
            model['Face'] = 'Denied'

    if 'Door Tampered' in percept:
        model['Alarm'] = 'Activated'
        if 'Owner Alarm Deactivated' in percept:
            model['Alarm'] = 'Deactivated'

    if 'PIN Entered Correctly' in percept or 'Correct Physical Key Used' in percept:
        model['Door'] = 'Open'
    elif 'PIN Entered Incorrectly' in percept or 'Incorrect Physical Key Used' in percept:
        if 'Owner Granted' in percept:
            model['Door'] = 'Open'
        elif 'Owner Denied' in percept:
            model['Door'] = 'Closed'

    if 'People Entered' in percept or 'People Exited' in percept:
        model['Door'] = 'Closed'

    # Update the model based on action
    if action == 'Camera Off':
        model['Camera'] = 'Off'
    if action == 'Camera On':
        model['Camera'] = 'On'
    if action == 'Door Open':
        model['Door'] = 'Open'
    if action == 'Keep Door Closed':
        model['Door'] = 'Closed'
    if action == 'Notify Owner':
        # Assume notifying the owner does not change the world state in this model
        pass
    if action == 'Alarm System Activate':
        model['Alarm'] = 'Activated'
    if action == 'Alarm Off':
        model['Alarm'] = 'Off'
    if action == 'Lock Door':
        model['Door'] = 'Closed'

    return model

def model_based_reflex_agent(percept):
    global world_state
    motion_detected = 'Motion Detected' in percept
    motion_not_detected = 'Motion Not Detected' in percept
    face_recognized = 'Face Recognized' in percept
    face_not_recognized = 'Face Not Recognized' in percept
    door_tampered = 'Door Tampered' in percept
    pin_entered_correctly = 'PIN Entered Correctly' in percept
    pin_entered_incorrectly = 'PIN Entered Incorrectly' in percept
    correct_key_used = 'Correct Physical Key Used' in percept
    incorrect_key_used = 'Incorrect Physical Key Used' in percept
    people_entered = 'People Entered' in percept
    people_exited = 'People Exited' in percept

    if motion_not_detected:
        action = 'Camera Off'
    elif motion_detected:
        if face_recognized:
            action = 'Door Open'
        elif face_not_recognized:
            if 'Owner Granted' in percept:
                action = 'Door Open'
            elif 'Owner Denied' in percept:
                action = 'Keep Door Closed'
            else:
                action = ('Keep Door Closed', 'Notify Owner')
    elif door_tampered:
        if 'Owner Alarm Deactivated' in percept:
            action = 'Alarm Off'
        else:
            action = ('Notify Owner', 'Alarm System Activate')
    elif pin_entered_correctly:
        action = 'Door Open'
    elif pin_entered_incorrectly:
        if 'Owner Granted' in percept:
            action = 'Door Open'
        elif 'Owner Denied' in percept:
            action = 'Keep Door Closed'
        else:
            action = ('Keep Door Closed', 'Notify Owner')
    elif correct_key_used:
        action = 'Door Open'
    elif incorrect_key_used:
        if 'Owner Granted' in percept:
            action = 'Door Open'
        elif 'Owner Denied' in percept:
            action = 'Keep Door Closed'
        else:
            action = ('Keep Door Closed', 'Notify Owner')
    elif people_entered or people_exited:
        action = 'Lock Door'

    world_state = update_state(action, percept, world_state)
    print('Perception: ' + str(percept))
    print('Action Performed: ' + str(action))
    print('World State: ' + str(world_state))
    return action
```

```python
# Let's simulate the agent
import random

# Simulating the first perception randomly
motion_detected = random.choice([True, False])
if motion_detected:
    face_recognized = random.choice([True, False])
    if face_recognized:
        percept = ('Motion Detected', 'Face Recognized')
    else:
        owner_granted = random.choice([True, False])
        percept = ('Motion Detected', 'Face Not Recognized')
        if owner_granted:
            percept += ('Owner Granted',)
        else:
            percept += ('Owner Denied',)
else:
    percept = ('Motion Not Detected',)

# To perceive environment repeatedly
while True:
    # The agent program is being called with current perception
    action = model_based_reflex_agent(percept)
    print('Action Performed: ' + str(action))

    # User control to allow next perception
    cmd = input('Type x to exit! or any key to continue.')
    if cmd == 'x': break

    # Simulating the next perception based on action
    if 'Camera Off' in action:
        motion_detected = random.choice([True, False])
        percept = ('Motion Not Detected',) if not motion_detected else ('Motion Detected',)
    elif 'Camera On' in action:
        face_recognized = random.choice([True, False])
        if face_recognized:
            percept = ('Motion Detected', 'Face Recognized')
        else:
            owner_granted = random.choice([True, False])
            percept = ('Motion Detected', 'Face Not Recognized')
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
    elif 'Door Open' in action:
        people_entered = random.choice([True, False])
        if people_entered:
            percept = ('People Entered',)
        else:
            people_exited = random.choice([True, False])
            if people_exited:
                percept = ('People Exited',)
    elif 'Alarm System Activate' in action:
        owner_alarm_deactivated = random.choice([True, False])
        percept = ('Door Tampered',)
        if owner_alarm_deactivated:
            percept += ('Owner Alarm Deactivated',)
    elif 'Keep Door Closed' in action or 'Notify Owner' in action:
        # Keep the same state for the next round
        pass

    # Simulating PIN or Physical Key entry
    pin_entered = random.choice([True, False])
    if pin_entered:
        pin_correct = random.choice([True, False])
        if pin_correct:
            percept = ('PIN Entered Correctly',)
        else:
            owner_granted = random.choice([True, False])
            percept = ('PIN Entered Incorrectly',)
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
    physical_key_used = random.choice([True, False])
    if physical_key_used:
        key_correct = random.choice([True, False])
        if key_correct:
            percept = ('Correct Physical Key Used',)
        else:
            owner_granted = random.choice([True, False])
            percept = ('Incorrect Physical Key Used',)
            if owner_granted:
                percept += ('Owner Granted',)
            else:
                percept += ('Owner Denied',)
```
